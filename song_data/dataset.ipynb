```python
import os
import urllib
from dotenv import load_dotenv
load_dotenv()

SPOTIFY_CLIENT_ID = os.getenv("SPOTIFY_CLIENT_ID")
SPOTIFY_CLIENT_SECRET = os.getenv("SPOTIFY_CLIENT_SECRET")
SPOTIFY_TOKEN_URL = "https://accounts.spotify.com/api/token"
```

```python
import requests
import base64

auth_str = f"{SPOTIFY_CLIENT_ID}:{SPOTIFY_CLIENT_SECRET}"
b64_auth_str = base64.b64encode(auth_str.encode()).decode()
response = requests.post(
    SPOTIFY_TOKEN_URL,
    data={"grant_type": "client_credentials"},
    headers={"Authorization": f"Basic {b64_auth_str}"}
)

token = response.json()['access_token']
```

```python
track_ids = df['id'].to_list()
headers = {"Authorization": f"Bearer {token}"}
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
import time 
import logging

logging.basicConfig(level=logging.INFO)

def fetch_track_batch(batch_ids, headers, retries=3):
    url = "https://api.spotify.com/v1/tracks"
    params = {"ids": ",".join(batch_ids)}

    for attempt in range(retries):
        try:
            response = requests.get(url, headers=headers, params=params)
            if response.status_code == 200:
                data = response.json()
                track_info = []
                for track in data.get('tracks', []):
                    if track:
                        track_info.append({
                            "id": track['id'],
                            "track_name": track['name'],
                            "artist_name": track['artists'][0]['name']
                        })
                return track_info
            elif response.status_code == 429:
                retry_after = int(response.headers.get("Retry-After", 1))
                logging.warning(f"Rate limited. Retrying after {retry_after}s")
                time.sleep(retry_after + 1)
            else:
                logging.warning(f"Failed batch {batch_ids[:3]}..., status {response.status_code}")
        except Exception as e:
            logging.error(f"Exception on batch {batch_ids[:3]}...: {e}")
        time.sleep(1 + attempt) 
    return []
```

```python
df['id'] = df['id'].astype(str).str.strip()
track_ids = df['id'].dropna().unique().tolist()
batches = [track_ids[i:i + 50] for i in range(0, len(track_ids), 50)]

track_info_list = []
with ThreadPoolExecutor(max_workers=5) as executor:
    futures = [executor.submit(fetch_track_batch, batch, headers) for batch in batches]
    for future in as_completed(futures):
        track_info_list.extend(future.result())

track_info_df = pd.DataFrame(track_info_list)
track_info_df['id'] = track_info_df['id'].astype(str).str.strip()
df = df.merge(track_info_df, on="id", how="left")

missing_count = df['track_name'].isna().sum()
total_count = len(df)
logging.info(f"Total tracks: {total_count}")
logging.info(f"Tracks with missing info: {missing_count}")
```
